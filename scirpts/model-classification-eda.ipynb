```python
def plot_train_test(train_dir, test_dir, save_name, view, alpha=0.7, colors=['#ADD8E6', '#FFA07A']):
    import os
    import matplotlib.pyplot as plt
    
    def count_images(directory):
        counts = {}
        for class_name in os.listdir(directory):
            class_path = os.path.join(directory, class_name)
            if os.path.isdir(class_path):
                counts[class_name] = len([
                    f for f in os.listdir(class_path)
                    if os.path.isfile(os.path.join(class_path, f))
                ])
        return counts
    
    train_counts = count_images(train_dir)
    test_counts = count_images(test_dir)
    all_classes = sorted(set(train_counts.keys()) | set(test_counts.keys()))
    
    train_stack = [train_counts.get(cls, 0) for cls in all_classes]
    test_stack = [test_counts.get(cls, 0) for cls in all_classes]
    
    colors = colors
    
    plt.figure(figsize=(6, 5))
    
    x_labels = ['Train', 'Test']
    x = range(len(x_labels))
    
    train_bottom = 0
    test_bottom = 0
    
    for i, cls in enumerate(all_classes):
        val = train_stack[i]
        plt.bar(0, val, bottom=train_bottom, color=colors[i], alpha=alpha, label=cls if i == 0 else "")
        plt.text(0, train_bottom + val / 2, str(val), ha='center', va='center', color='black', fontsize=18)
        train_bottom += val
    
    for i, cls in enumerate(all_classes):
        val = test_stack[i]
        plt.bar(1, val, bottom=test_bottom, color=colors[i], alpha=alpha, label=cls if i == 1 else "")
        plt.text(1, test_bottom + val / 2, str(val), ha='center', va='center', color='black', fontsize=18)
        test_bottom += val
    
    plt.xticks(x, x_labels, fontsize=20)
    plt.yticks(fontsize=16)
    plt.ylabel("Number of Images", fontsize=20)
    plt.title(f'{view}', fontsize=20)
    plt.legend(['Moderate Dementia', 'Non-Dementia'], fontsize=14)
    plt.tight_layout()
    # plt.savefig(f'../plots/outputs/{save_name}.pdf', bbox_inches='tight')
    plt.show()
```

```python
import os
import matplotlib.pyplot as plt

def count_images(directory):
    counts = {}
    for class_name in os.listdir(directory):
        class_path = os.path.join(directory, class_name)
        if os.path.isdir(class_path):
            counts[class_name] = len([
                f for f in os.listdir(class_path)
                if os.path.isfile(os.path.join(class_path, f))
            ])
    return counts


def plot_train_test_all(datasets, save_name, alpha=0.7):
    fig, axes = plt.subplots(1, 4, figsize=(22, 5))  # 1 row, 4 columns
    subplot_labels = ['(a)', '(b)', '(c)', '(d)']
    legend_labels = ['Moderate Dementia', 'Non-Dementia']  # fixed legend

    for idx, dataset in enumerate(datasets):
        ax = axes[idx]

        train_counts = count_images(dataset["train_dir"])
        test_counts = count_images(dataset["test_dir"])
        all_classes = sorted(set(train_counts.keys()) | set(test_counts.keys()))

        # Assign colors per dataset
        if "colors" in dataset:
            plot_colors = dataset["colors"]
        else:
            plot_colors = ['#ADD8E6', '#FFA07A']  # default for first 3 plots

        class_to_color = {cls: plot_colors[i] for i, cls in enumerate(all_classes)}

        train_stack = [train_counts.get(cls, 0) for cls in all_classes]
        test_stack = [test_counts.get(cls, 0) for cls in all_classes]

        x_labels = ['Train', 'Test']
        x = range(len(x_labels))
        train_bottom, test_bottom = 0, 0

        for cls in all_classes:
            c = class_to_color[cls]

            # Train bar
            val = train_counts.get(cls, 0)
            ax.bar(0, val, bottom=train_bottom, color=c, alpha=alpha)
            if val > 0:
                ax.text(0, train_bottom + val / 2, str(val),
                        ha='center', va='center', color='black', fontsize=20)
            train_bottom += val

            # Test bar
            val = test_counts.get(cls, 0)
            ax.bar(1, val, bottom=test_bottom, color=c, alpha=alpha)
            if val > 0:
                ax.text(1, test_bottom + val / 2, str(val),
                        ha='center', va='center', color='black', fontsize=20)
            test_bottom += val

        ax.set_xticks(x, x_labels, fontsize=22)
        ax.set_ylabel("Number of Images", fontsize=22)
        ax.set_title(dataset["view"], fontsize=22)
        ax.tick_params(axis='y', labelsize=18)

        # Bold legend with explicit colors
        handles = [plt.Rectangle((0,0),1,1, color=class_to_color[cls]) for cls in all_classes]
        ax.legend(handles, legend_labels, fontsize=12)

        # Add subplot label at bottom center
        ax.text(0.5, -0.25, subplot_labels[idx], fontsize=20, fontweight='bold',
                ha='center', va='center', transform=ax.transAxes)

    plt.tight_layout()
    plt.savefig(f'../plots/outputs/{save_name}.pdf', bbox_inches='tight')
    plt.show()



# ==== Usage ====
datasets = [
    {
        "train_dir": r"D:\Research\Vision\mra\alz-cq-cnn\datasets\OASIS-2\preprocessed_balanced\train\axial",
        "test_dir": r"D:\Research\Vision\mra\alz-cq-cnn\datasets\OASIS-2\preprocessed_balanced\test\axial",
        "view": "Axial"
    },
    {
        "train_dir": r"D:\Research\Vision\mra\alz-cq-cnn\datasets\OASIS-2\preprocessed_balanced\train\coronal",
        "test_dir": r"D:\Research\Vision\mra\alz-cq-cnn\datasets\OASIS-2\preprocessed_balanced\test\coronal",
        "view": "Coronal"
    },
    {
        "train_dir": r"D:\Research\Vision\mra\alz-cq-cnn\datasets\OASIS-2\preprocessed_balanced\train\sagittal",
        "test_dir": r"D:\Research\Vision\mra\alz-cq-cnn\datasets\OASIS-2\preprocessed_balanced\test\sagittal",
        "view": "Sagittal"
    },
    {
        "train_dir": r"D:\Research\Vision\mra\alz-cq-cnn\datasets\OASIS-2\preprocessed_balanced\train\3-plane",
        "test_dir": r"D:\Research\Vision\mra\alz-cq-cnn\datasets\OASIS-2\preprocessed_balanced\test\3-plane",
        "view": "3-Plane",
        "colors": ['#7AAEDD', '#E67357']
    }
]

plot_train_test_all(datasets, save_name="oasis-2-dataset")
```
