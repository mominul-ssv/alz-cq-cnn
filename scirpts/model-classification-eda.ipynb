```python
import pandas as pd
import numpy as np
from math import pi
import matplotlib.pyplot as plt
```

```python
# List of file paths
file_paths = [
    r'D:\Research\Vision\mra\root\models\classification\classical\oasis-2_axial.csv',
    r'D:\Research\Vision\mra\root\models\classification\classical\oasis-2_axial-ss.csv',
    r'D:\Research\Vision\mra\root\models\classification\classical\oasis-2_coronal.csv',
    r'D:\Research\Vision\mra\root\models\classification\classical\oasis-2_coronal-ss.csv',
    r'D:\Research\Vision\mra\root\models\classification\classical\oasis-2_sagittal.csv',
    r'D:\Research\Vision\mra\root\models\classification\classical\oasis-2_sagittal-ss.csv',
    r'D:\Research\Vision\mra\root\models\classification\classical\oasis-2_3-plane.csv',
    r'D:\Research\Vision\mra\root\models\classification\classical\oasis-2_3-plane-ss.csv'
]

# Simple labels corresponding to each dataset
simple_labels = [
    'Classical | Axial', 'Classical | Axial ($\Xi$)', 'Classical | Coronal', 'Classical | Coronal ($\Xi$)', 
    'Classical | Sagittal', 'Classical | Sagittal ($\Xi$)', 'Classical | 3-Plane', 'Classical | 3-Plane ($\Xi$)'
]

# Set up the plot with 2 subplots (one for train accuracy, one for validation accuracy)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 3))

# Iterate through each file and plot both train and validation accuracy
for file_path, label in zip(file_paths, simple_labels):
    # Load the CSV file into a pandas DataFrame
    df = pd.read_csv(file_path)
    
    # Plot train accuracy on the first subplot (ax1) and label only once
    ax1.plot(df['epoch'], df['train_accuracy'], label=label, marker='o')
    
    # Plot validation accuracy on the second subplot (ax2) and label only once
    ax2.plot(df['epoch'], df['val_accuracy'], label=label, marker='o')

# Customize the first plot (Train Accuracy)
ax1.set_xlabel('Epoch', fontsize=13)  # Increased font size
ax1.set_ylabel('Train Accuracy (%)  \n [CNN | Classical]', fontsize=13)  # Increased font size
ax1.grid(True)

# Customize the second plot (Validation Accuracy)
ax2.set_xlabel('Epoch', fontsize=13)  # Increased font size
ax2.set_ylabel('Validation Accuracy (%) \n [CNN | Classical]', fontsize=13)  # Increased font size
ax2.grid(True)
ax2.legend(loc='upper left', bbox_to_anchor=(0.695, 0.766), fontsize=9)

# Adjust layout to prevent clipping of labels
plt.tight_layout()
plt.savefig(f'../plots/outputs/train-val-classical.pdf', bbox_inches='tight')
plt.show()
```

```python
# Data for 2-qubit models (first table) without the ± standard deviation
precision_2q = [87.77, 89.55, 94.71, 69.31, 95.69, 70.68, 90.34, 67.21]
f1_score_2q = [93.45, 93.42, 97.27, 81.43, 97.75, 80.88, 92.46, 75.27]
specificity_2q = [98.43, 98.53, 99.30, 94.05, 99.41, 94.04, 98.61, 95.70]
accuracy_2q = [98.40, 98.51, 99.37, 94.71, 99.48, 93.70, 98.23, 93.50]
training_time_2q = [
    8*60 + 20,  # 00:08:20 -> 500 seconds
    7*60 + 17,  # 00:07:17 -> 437 seconds
    4*60 + 24,  # 00:04:24 -> 264 seconds
    4*60 + 14,  # 00:04:14 -> 254 seconds
    4*60 + 28,  # 00:04:28 -> 268 seconds
    4*60 + 51,  # 00:04:51 -> 291 seconds
    25*60 + 18, # 00:25:18 -> 1518 seconds
    24*60 + 53  # 00:24:53 -> 1493 seconds
]

# Data for 3-qubit models (second table) without the ± standard deviation
precision_3q = [90.69, 78.83, 91.86, 71.96, 94.92, 78.14, 90.23, 83.19]
f1_score_3q = [95.12, 87.40, 95.75, 81.23, 90.94, 86.76, 94.85, 89.45]
specificity_3q = [98.72, 96.86, 98.95, 92.83, 99.29, 95.75, 98.64, 97.55]
accuracy_3q = [98.86, 97.01, 98.96, 93.62, 98.11, 96.22, 98.79, 97.50]
training_time_3q = [
    22*60 + 40, # 00:22:40 -> 1360 seconds
    18*60 + 43, # 00:18:43 -> 1123 seconds
    10*60 + 52, # 00:10:52 -> 652 seconds
    10*60 + 11, # 00:10:11 -> 611 seconds
    10*60 + 15, # 00:10:15 -> 615 seconds
    10*60 + 7,  # 00:10:07 -> 607 seconds
    83*60 + 5,  # 01:23:05 -> 4985 seconds
    80*60 + 55  # 01:20:55 -> 4855 seconds
]

# Metric names (including Training Time)
categories = ['PRE', 'F1', 'SPEC', 'ACC', 'T. Time']

# Data for radar plot (combine the metrics, including Training Time)
values_2q = [
    [precision_2q[i], f1_score_2q[i], specificity_2q[i], accuracy_2q[i], training_time_2q[i]] 
    for i in range(len(precision_2q))
]

values_3q = [
    [precision_3q[i], f1_score_3q[i], specificity_3q[i], accuracy_3q[i], training_time_3q[i]] 
    for i in range(len(precision_3q))
]

# Number of variables
N = len(categories)

# Compute angle for each axis of the radar plot
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]  # Complete the loop

# Find the maximum training time across both 2-qubit and 3-qubit models
max_training_time = max(max(training_time_2q), max(training_time_3q))

# Normalize the training time to be between 0 and 100 (out of the maximum)
training_time_2q_normalized = np.array(training_time_2q) / max_training_time * 100
training_time_3q_normalized = np.array(training_time_3q) / max_training_time * 100

# Replace training time values with normalized values
for i in range(len(values_2q)):
    values_2q[i][4] = training_time_2q_normalized[i]
    values_3q[i][4] = training_time_3q_normalized[i]

# Custom names for the models (Alpha vs Beta format)
model_names = [
    '$\\alpha_1$ vs $\\beta_1$', '$\\alpha_2$ vs $\\beta_2$', '$\\alpha_3$ vs $\\beta_3$', '$\\alpha_4$ vs $\\beta_4$',
    '$\\alpha_5$ vs $\\beta_5$', '$\\alpha_6$ vs $\\beta_6$', '$\\alpha_7$ vs $\\beta_7$', '$\\alpha_8$ vs $\\beta_8$'
]

# Custom colors for even and odd models (similar tone for corresponding models)
odd_color_2q = '#2ca02c'   # blue for odd 2-qubit models
odd_color_3q = '#e377c2'   # orange for odd 3-qubit models
even_color_2q = '#1f77b4'  # green shade for even 2-qubit models
even_color_3q = '#ff7f0e'  # pink shade for even 3-qubit models

# Create a big figure with 2 rows and 4 columns of subplots
fig, axes = plt.subplots(2, 4, figsize=(18, 9), dpi=100, subplot_kw=dict(polar=True))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Create 8 separate radar plots, one for each model
for i in range(8):
    ax = axes[i]
    
    # Decide color based on the index (even or odd models)
    if i % 2 == 0:  # Even models (0, 2, 4, 6)
        color_2q = even_color_2q
        color_3q = even_color_3q
    else:  # Odd models (1, 3, 5, 7)
        color_2q = odd_color_2q
        color_3q = odd_color_3q

    # Plot the 2-qubit model data (close the loop)
    ax.plot(angles, values_2q[i] + [values_2q[i][0]], linewidth=3, linestyle='solid', color=color_2q, label="2-qubit")
    ax.fill(angles, values_2q[i] + [values_2q[i][0]], alpha=0.25, color=color_2q)

    # Plot the 3-qubit model data (close the loop)
    ax.plot(angles, values_3q[i] + [values_3q[i][0]], linewidth=3, linestyle='solid', color=color_3q, label="3-qubit")
    ax.fill(angles, values_3q[i] + [values_3q[i][0]], alpha=0.25, color=color_3q)

    # Add labels to the plot
    ax.set_yticklabels([])
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontweight='bold', fontsize=12)

    # Set the title for each subplot using the custom names
    ax.set_title(model_names[i], size=16, fontweight='bold')

    # Add a legend
    ax.legend(loc='lower left', fontsize=10, bbox_to_anchor=(-0.15, -0.15))

# Adjust the layout to fit the subplots
plt.tight_layout()
plt.savefig(f'../plots/outputs/radar-plot.pdf', bbox_inches='tight')
plt.show()
```
