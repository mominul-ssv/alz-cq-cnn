# t-SNE

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from PIL import Image
```

### MNIST

```python
def load_images_from_folder(folder_path):
    X = []
    y = []
    class_folders = [d for d in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, d))]
    for class_label in class_folders:
        class_path = os.path.join(folder_path, class_label)
        for filename in os.listdir(class_path):
            if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp')):
                img_path = os.path.join(class_path, filename)
                img = Image.open(img_path).convert('L')  # grayscale
                img = img.resize((28,28))  # resize if needed (MNIST default)
                img_array = np.array(img).flatten()
                X.append(img_array)
                y.append(int(class_label))  # assuming folder name is class label as int
    X = np.array(X)
    y = np.array(y)
    return X, y

def plot_tsne(ax, X, y, title, cmap, labels):
    tsne = TSNE(n_components=2, random_state=42)
    X_embedded = tsne.fit_transform(X)
    scatter = ax.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap=cmap, s=10, alpha=0.7)
    ax.set_title(title, fontsize=20)
    ax.set_xticks([])
    ax.set_yticks([])

    handles, _ = scatter.legend_elements()
    ax.legend(handles, labels, fontsize=14, loc='upper left', bbox_to_anchor=(0, 1), frameon=True)
    
    return scatter

folders = {
    '0v1': r'D:\Research\Vision\mra\alz-cq-cnn\datasets\MNIST\0v1\train',
    '2v3': r'D:\Research\Vision\mra\alz-cq-cnn\datasets\MNIST\2v3\train',
    '4v5': r'D:\Research\Vision\mra\alz-cq-cnn\datasets\MNIST\4v5\train',
}

cmaps = ['Set1', 'Accent', 'tab10']

fig, axes = plt.subplots(1, 3, figsize=(16, 8))

label_map = [['$\\mathdefault{0}$', '$\\mathdefault{1}$'], 
             ['$\\mathdefault{2}$', '$\\mathdefault{3}$'], 
             ['$\\mathdefault{4}$', '$\\mathdefault{5}$']]

for ax, (label, folder), cmap, l_map in zip(axes, folders.items(), cmaps, label_map):
    print(f'Loading {label} data...')
    X, y = load_images_from_folder(folder)
    print(f'{label}: {X.shape[0]} samples loaded')
    plot_tsne(ax, X, y, f't-SNE of {label}', cmap=cmap, labels=l_map)

plt.subplots_adjust(wspace=0.05) 
plt.savefig(f'../plots/outputs/control-mnist-t-sne.png', bbox_inches='tight')
plt.show()
```

### OASIS-2

```python
def load_images_from_folder(folder_path):
    X = []
    y = []
    class_label_map = {'nondemented': 0, 'moderate_dementia': 1}
    class_folders = [d for d in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, d))]
    
    for class_label in class_folders:
        class_path = os.path.join(folder_path, class_label)
        for filename in os.listdir(class_path):
            if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp')):
                img_path = os.path.join(class_path, filename)
                img = Image.open(img_path).convert('L')  # grayscale
                img = img.resize((128,128))  # Resizing to 128x128 for visualization
                img_array = np.array(img).flatten()
                X.append(img_array)
                y.append(class_label_map.get(class_label.lower(), -1))  # Safely map label, -1 if not mapped
    X = np.array(X)
    y = np.array(y)
    return X, y

def plot_tsne(ax, X, y, title, cmap):
    tsne = TSNE(n_components=2, random_state=42)
    X_embedded = tsne.fit_transform(X)
    scatter = ax.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap=cmap, s=10, alpha=0.7)
    ax.set_title(title, fontsize=20)
    ax.set_xticks([])
    ax.set_yticks([])

    # Define proper labels
    label_map = {0: 'Non-Dementia', 1: 'Moderate Dementia'}
    handles, _ = scatter.legend_elements()
    labels = [label_map[label] for label in np.unique(y)]
    
    ax.legend(handles, labels, fontsize=14, loc='upper right', bbox_to_anchor=(1, 1), frameon=True)
    return scatter

folders = {
    'Axial': r'D:\Research\Vision\mra\alz-cq-cnn\datasets\OASIS-2\preprocessed_balanced\train\axial',
    'Coronal': r'D:\Research\Vision\mra\alz-cq-cnn\datasets\OASIS-2\preprocessed_balanced\train\coronal',
    'Sagittal': r'D:\Research\Vision\mra\alz-cq-cnn\datasets\OASIS-2\preprocessed_balanced\train\sagittal',
}

cmaps = ['Set1', 'Accent', 'tab10']

fig, axes = plt.subplots(1, 3, figsize=(16, 8))

for ax, (label, folder), cmap in zip(axes, folders.items(), cmaps):
    print(f'Loading {label} data from {folder}...')
    X, y = load_images_from_folder(folder)
    print(f'{label}: {X.shape[0]} samples loaded')
    plot_tsne(ax, X, y, f't-SNE of {label} MRI', cmap=cmap)

plt.subplots_adjust(wspace=0.05)
plt.savefig(f'../plots/outputs/control-oasis-2-t-sne.png', bbox_inches='tight')
plt.show()
```
